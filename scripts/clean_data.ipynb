# Standardizing column extra spaces

```python
df["name"]      = df["name"].str.strip()
df["brand"]     = df["brand"].str.strip()
df["segment"]   = df["segment"].str.strip()
df["fuel_type"] = df["fuel_type"].str.strip()
```

# filling Missing value

```python
df["mileage_kmpl"] = df["mileage_kmpl"].fillna(0)
df["range_km"]     = df["range_km"].fillna(0)
df["battery_kwh"]  = df["battery_kwh"].fillna(0)
df["engine_cc"]    = df["engine_cc"].fillna(0)
```

```python
def price_band(p):
    if p < 80000:   return "Budget (Under 80K)"
    elif p < 150000: return "Mid-range (80K–1.5L)"
    elif p < 250000: return "Premium (1.5L–2.5L)"
    else:            return "Luxury (Above 2.5L)"

df["price_band"] = df["price_inr"].apply(price_band)
```

```python
# Add value score — power per lakh spent
df["value_score"] = round(df["power_bhp"] / (df["price_inr"] / 100000), 2)
```

```python
# Add efficiency label
df["efficiency"] = df.apply(
    lambda r: r["range_km"] if r["is_ev"] else r["mileage_kmpl"], axis=1
)
```

```python
df.to_csv(f"{CLEAN}/bikes_clean.csv", index=False)
```

# Cleaning Sales_monthly 

```python
sales = pd.read_csv(f"{RAW}/sales_monthly.csv")
```

```python
sales.head()
```

```python
sales["month_date"] = pd.to_datetime(sales["month_date"])
sales["year"]       = sales["year"].astype(int)
sales["month"]      = sales["month"].astype(int)
sales["units_sold"] = sales["units_sold"].fillna(0).astype(int)
sales["name"]       = sales["name"].str.strip()
```

```python
# Add quarter column
sales["quarter"] = sales["month"].apply(
    lambda m: f"Q{(m-1)//3+1}"
)
```

```python
# Add festive flag
sales["is_festive"] = sales["month"].isin([9, 10, 11])
```

```python
# Add COVID flag
sales["is_covid"] = (
    ((sales["year"] == 2020) & (sales["month"].isin([3,4,5,6]))) |
    ((sales["year"] == 2021) & (sales["month"].isin([4,5])))
)
```

```python
sales.to_csv(f"{CLEAN}/sales_clean.csv", index=False)
```

# Cleaning EV STATE WISE

```python
states = pd.read_csv(f"{RAW}/ev_state_wise.csv")
```

```python
states["month_date"]      = pd.to_datetime(states["month_date"])
states["state"]           = states["state"].str.strip()
states["ev_registrations"]= states["ev_registrations"].fillna(0).astype(int)
states["charging_stations"]= states["charging_stations"].fillna(0).astype(int)
```

```python
# Add region mapping
region_map = {
    "Maharashtra":"West","Gujarat":"West","Rajasthan":"West",
    "Delhi":"North","Uttar Pradesh":"North","Punjab":"North","Haryana":"North","Himachal Pradesh":"North",
    "Karnataka":"South","Tamil Nadu":"South","Telangana":"South","Andhra Pradesh":"South","Kerala":"South",
    "West Bengal":"East","Odisha":"East","Bihar":"East","Jharkhand":"East","Assam":"East",
    "Madhya Pradesh":"Central","Chhattisgarh":"Central"
}
```

```python
states["region"] = states["state"].map(region_map)
```

```python
states.to_csv(f"{CLEAN}/ev_states_clean.csv", index=False)
```

# Cleaning CUSTOMER REVIEWS

```python
reviews = pd.read_csv(f"{RAW}/customer_reviews.csv")
```

```python
reviews["bike_name"] = reviews["bike_name"].str.strip()
reviews["brand"]     = reviews["brand"].str.strip()
reviews["rating"]    = reviews["rating"].clip(1, 5)
```

```python
# Sentiment bucket
reviews["sentiment"] = reviews["rating"].apply(
    lambda r: "Positive" if r >= 4 else ("Neutral" if r >= 3 else "Negative")
)
```

```python
reviews.to_csv(f"{CLEAN}/reviews_clean.csv", index=False)
```
